```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Setup Data Categories and Plot Titles
classes = np.arange(10)  # Classes 0-9
titles = ['iid', 'dir', 'prob', 'qty', 'noise', 'qs'] # Titles from your image

# 2. Initialize the Figure (2 rows, 3 columns)
# sharex/sharey ensures axes are aligned and clean like in your image
fig, axes = plt.subplots(2, 3, figsize=(12, 6), sharex=True, sharey=True)
axes = axes.flatten() # Flatten 2D array of axes to 1D for easy looping

# 3. Define Colors (Blue, Orange, Red)
colors = ['#4e79a7', '#f28e2b', '#e15759']
labels = ['$u_1$', '$u_2$', '$u_3$']

# 4. Loop through each subplot to create the charts
for i, ax in enumerate(axes):
    # --- Generate Dummy Data (Replace this with your real data) ---
    # Create random data for 3 users across 10 classes
    # Shape: (3, 10) -> 3 users, 10 values each
    data = np.random.randint(5, 50, size=(3, 10))
    
    # Specific adjustment to mimic the 'iid' (uniform) look
    if titles[i] == 'iid':
        data = np.random.randint(15, 16, size=(3, 10))

    # --- Plotting Stacked Bars ---
    bottom = np.zeros(10) # Start the stack at 0
    for j, user_data in enumerate(data):
        # 'bottom' parameter is key for stacking
        ax.bar(classes, user_data, bottom=bottom, 
               color=colors[j], label=labels[j] if i == 5 else "")
        # Update bottom for the next layer
        bottom += user_data

    # --- Formatting ---
    ax.set_title(titles[i], fontsize=14)
    ax.set_xticks(classes)
    ax.grid(axis='y', linestyle='--', alpha=0.5)

    # Add X/Y Labels only to outer plots (like in your image)
    if i >= 3: 
        ax.set_xlabel('Classes (0-9)', fontsize=12)
    if i % 3 == 0: 
        ax.set_ylabel('Data Size', fontsize=12)

# 5. Add Legend to the last plot
axes[5].legend()

# 6. Adjust Layout and Save
plt.tight_layout()
plt.savefig('recreated_graph.png')
plt.show()
```
